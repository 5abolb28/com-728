```python
# os module in python provides functions for interacting with the operating system
import os

from load_functions import *
from process_functions import *
from utilise_functions import *
from visualise_functions import *
from tui_functions import *

'''
Main Function
'''
def run():
    data_loaded = False #Flag to check if data is loaded
    while data_loaded == False: #Loop until data is loaded
        print("Welcome!!")
        print("Our current directory is:", os.getcwd())
        df_path = input("Please enter path to the dataset 'CarPrice.csv' (Full or Relative) and press enter: ")
        try: #Check if the path is valid
            df = load_data_from_path(df_path)
            print("Data Loaded Successfully")
            data_loaded = True
        except: #If the data is not loaded
            print("Incorrect Path Entered! Try Again!!")
            
    program_run = True #Flag to check if the program is running
    while program_run == True: #Loop until the program is running
        
        if data_loaded == True: #If data is loaded
            user_select = display_main_option() #To display main option
            if (user_select > 0 and user_select < 4): #If the input is valid
                print('Loading Sub-Menu....') #Display loading message
                
                if user_select == 1: #If the user selects to process data
                    user_sub_select = display_process_option() #To display process option
                    if user_sub_select == 1: #If the user selects to retrieve a record for an individual car by id
                        car_id = int(input("Enter Car ID: "))
                        result = get_record_by_id(df, car_id)
                        if type(result) == int: #If the car id is not found
                            print("Invalid Car ID Entered!! Try again.")
                        else:
                            print(result)
                    elif user_sub_select == 2: #if the user selects to retrieve all cars for a specified cylinder number
                        cyl_no = int(input("Enter Cylinder Number: "))
                        result = get_cars_by_cyl(df, cyl_no)
                        if type(result) == int: #If the cylinder number is not found
                            print("Invalid Cylinder Number Entered!! Try again.")
                        else:
                            print(result)
                    elif user_sub_select == 3: #If the user selects to retrieve all cars in the specified car body
                        car_body = input("Enter Car Body ('sedan', 'hatchback', 'wagon', 'hardtop', 'convertible'):\n")
                        result = get_cars_by_carbody(df, car_body)
                        if type(result) == int: #If the car body is not found
                            print("Invalid Car Body Entered!! Try again.")
                        else:
                            print(result)
                    elif user_sub_select == 4: #If the user selects to retrieve a specific number of columns related to an individual car by id
                        car_id = int(input("Enter Car ID: "))
                        col_names_list = input("Enter Column Names separated by comma ('CarName', 'fueltype', 'doornumber', 'carbody',\n'drivewheel', 'enginelocation', 'wheelbase', 'carlength',\n'carwidth', 'carheight', 'curbweight', 'enginetype',\n'cylindernumber', 'enginesize', 'fuelsystem', 'horsepower',\n'citympg', 'highwaympg', 'price'):\n")
                        col_names_list = [i.strip() for i in col_names_list.split(',')]
                        result = get_record_by_id_columns(df, car_id, col_names_list)
                        print(result)
                    else: #If the user selects an invalid menu option
                        display_error() #Display error message
                        
                elif user_select == 2: #If the user selects to utilise data
                    user_sub_select = display_utilise_option() #Display utilise option
                    if user_sub_select == 1: #If the user selects to retrieve the car names alphabetically
                        result = get_car_name_alpha(df)
                        print(result)
                    elif user_sub_select == 2: #If the user selects to retrieve summary of sales (total car price) for each car body
                        result = get_car_body_total_price(df)
                        print(result)
                    elif user_sub_select == 3: #If the user selects to retrieve the top 5 car sale by price (the most expensive) and per car body
                        result = get_car_body_total_top5(df)
                        print(result)
                    elif user_sub_select == 4: #If the user selects to retrieve the details of cars based on your own selection
                        col_name = input("Enter a Column Name ('CarName', 'fueltype', 'doornumber', 'carbody',\n'drivewheel', 'enginelocation', 'wheelbase', 'carlength',\n'carwidth', 'carheight', 'curbweight', 'enginetype',\n'cylindernumber', 'enginesize', 'fuelsystem', 'horsepower',\n'citympg', 'highwaympg', 'price'):\n")
                        print('You entered', col_name)
                        col_value = input(f"Enter a value from column '{col_name}' to filter dataset:\n")
                        result = get_selected_data(df, col_name, col_value)
                        print(result)
                    else: #If the user selects an invalid menu option
                        display_error() #Display error message
                        
                elif user_select == 3: #If the user selects to visualise data
                    user_sub_select = display_visualise_option() #Display visualise option
                    if user_sub_select == 1: #If the user selects to display the number of cars per fuel system using a bar chart
                        get_cars_fuel_system(df)
                    elif user_sub_select == 2: #If the user selects to display the horsepower of the top 5 car sale by price (the cheapest) using a subplot
                        get_hp_top5(df)
                    elif user_sub_select == 3: #If the user selects to display the price of the top 5 car sale by price (the most expensive) using a suitable visualisation3
                        
                        col_name = input("Enter a Column Name ('CarName', 'fueltype', 'doornumber', 'carbody',\n'drivewheel', 'enginelocation', 'wheelbase', 'carlength',\n'carwidth', 'carheight', 'curbweight', 'enginetype',\n'cylindernumber', 'enginesize', 'fuelsystem', 'horsepower',\n'citympg', 'highwaympg', 'price'):\n")
                        print('Loading an appropriate Visualisation...')
                        get_selected_viz(df, col_name)
                    else: #If the user selects an invalid menu option
                        display_error() #Display error message
                        
                        
            elif user_select == 4: #If the user selects to exit the program
                print("Thanks! Terminating the Program!!")
                program_run = False #Set the flag to false to terminate the program
                
            else: #If the user selects an invalid menu option
                display_error() 
                
                
                
                
if __name__ == "__main__": #If the program is run directly
    run()
    
    
    
```
